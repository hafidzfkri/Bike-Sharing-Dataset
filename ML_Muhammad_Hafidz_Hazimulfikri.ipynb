## Landasan Teori

- **Analisis data** adalah proses mengolah data mentah menjadi informasi bermakna (pola, tren, hubungan) untuk mendukung pengambilan keputusan berbasis data.

- **Data cleaning** dan **data wrangling** penting sebelum analisis lanjutan agar data konsisten, bebas missing/duplikasi, dan siap dianalisis (contoh: konversi tipe tanggal, mapping kategori, agregasi).

## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Data Wrangling

### Gathering Data

```python
day_df = pd.read_csv("day.csv")
hour_df = pd.read_csv("hour.csv")

day_df.head()
hour_df.head()
```

**Insight:**
- Dataset day.csv berisi data harian 731 baris dan hour.csv berisi data per jam 17.379 baris.
- Target utama analisis adalah cnt (total penyewaan), dengan komponen casual dan registered.

### Assessing Data

```python
day_df.info()
hour_df.info()

day_df.isna().sum()
hour_df.isna().sum()

day_df.duplicated().sum()
hour_df.duplicated().sum()

day_df.describe()
```

**Insight:**
- Tidak ditemukan missing value dan duplikasi pada kedua dataset.
- Beberapa kolom bersifat kategori tetapi disimpan dalam bentuk angka (contoh: season, weathersit, weekday), sehingga perlu dibuat mapping agar lebih mudah dibaca.


### Cleaning Data

```python
day_df["dteday"] = pd.to_datetime(day_df["dteday"])
hour_df["dteday"] = pd.to_datetime(hour_df["dteday"])

season_map = {1:"Spring", 2:"Summer", 3:"Fall", 4:"Winter"}
weather_map = {
    1:"Clear/Few clouds",
    2:"Mist/Cloudy",
    3:"Light snow/rain",
    4:"Heavy rain/snow"
}

day_df["season_name"] = day_df["season"].map(season_map)
day_df["weather_name"] = day_df["weathersit"].map(weather_map)

hour_df["season_name"] = hour_df["season"].map(season_map)
hour_df["weather_name"] = hour_df["weathersit"].map(weather_map)

day_df["year"] = day_df["yr"].map({0:2011, 1:2012})
hour_df["year"] = hour_df["yr"].map({0:2011, 1:2012})

day_df.head()
```

**Insight:**
- Konversi dteday ke format datetime membuat analisis tren waktu lebih rapi.
- Mapping kategori (season & weather) membantu interpretasi hasil EDA dan visualisasi.
- Penambahan kolom year memudahkan perbandingan performa antar tahun (2011 vs 2012).

## Exploratory Data Analysis (EDA)

### Explore ...

```python
corr = (day_df[["temp","atemp","hum","windspeed","cnt"]]
        .corr(numeric_only=True)["cnt"]
        .drop("cnt")
        .sort_values(ascending=False))

display(corr)
```

**Insight:**

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
# Rata-rata rental per jam (overall)
avg_hour = hour_df.groupby("hr")["cnt"].mean().sort_values(ascending=False)
display(avg_hour.head(10))

# Workingday vs non-workingday
avg_hour_wd = hour_df[hour_df["workingday"]==1].groupby("hr")["cnt"].mean().sort_values(ascending=False)
avg_hour_nwd = hour_df[hour_df["workingday"]==0].groupby("hr")["cnt"].mean().sort_values(ascending=False)

print("Top 5 peak hours (overall):")
display(avg_hour.head(5))

print("\nTop 5 peak hours (working day):")
display(avg_hour_wd.head(5))

print("\nTop 5 peak hours (non-working day):")
display(avg_hour_nwd.head(5))

#VISUALISASI
# Garis rata-rata per jam (overall)
avg_hour_plot = hour_df.groupby("hr")["cnt"].mean().reset_index()
plt.figure(figsize=(10,4))
sns.lineplot(data=avg_hour_plot, x="hr", y="cnt", marker="o")
plt.title("Rata-rata Rental per Jam (Overall)")
plt.xlabel("Jam")
plt.ylabel("Rata-rata cnt")
plt.xticks(range(0,24))
plt.show()

# Workingday vs Non-workingday
pivot = hour_df.pivot_table(values="cnt", index="hr", columns="workingday", aggfunc="mean").reset_index()
pivot.columns = ["hr", "Non-workingday", "Workingday"]

plt.figure(figsize=(10,4))
plt.plot(pivot["hr"], pivot["Non-workingday"], marker="o", label="Non-workingday")
plt.plot(pivot["hr"], pivot["Workingday"], marker="o", label="Workingday")
plt.title("Rata-rata Rental per Jam: Workingday vs Non-workingday")
plt.xlabel("Jam")
plt.ylabel("Rata-rata cnt")
plt.xticks(range(0,24))
plt.legend()
plt.show()

```

**Insight:**
- Jam puncak biasanya muncul di 08.00 dan 17.0018.00
- Pada workingday, puncak lebih tajam (mobilitas rutin) sedangkan pada non-workingday, puncak cenderung lebih menyebar (rekreasi siangsore).

### Pertanyaan 2:

```python
weather_mean = day_df.groupby("weather_name")["cnt"].mean().sort_values(ascending=False)
season_mean  = day_df.groupby("season_name")["cnt"].mean().sort_values(ascending=False)

print("Rata-rata cnt berdasarkan cuaca:")
display(weather_mean)

print("\nRata-rata cnt berdasarkan musim:")
display(season_mean)

#VISUALISASI
plt.figure(figsize=(10,4))
sns.barplot(x=weather_mean.index, y=weather_mean.values)
plt.title("Rata-rata Rental Harian berdasarkan Kondisi Cuaca")
plt.xlabel("Kondisi Cuaca")
plt.ylabel("Rata-rata cnt")
plt.xticks(rotation=15)
plt.show()

plt.figure(figsize=(8,4))
sns.barplot(x=season_mean.index, y=season_mean.values)
plt.title("Rata-rata Rental Harian per Musim")
plt.xlabel("Musim")
plt.ylabel("Rata-rata cnt")
plt.show()



```

**Insight:**
- Cuaca lebih cerah (Clear/Few clouds) cenderung menghasilkan rental tertinggi, sedangkan cuaca buruk (mis. hujan/salju) menurunkan demand.
- Musim tertentu (biasanya Fall/Summer) memiliki demand lebih tinggi dibanding musim lain.
